## 판다스 데이터 가공하기

### 칼럼(변수) 삭제/생성하기

```python
dust= pd.read_excel('dust1.xlsx')
dust
```

```python
dust.drop(['지역', '망', '측정소코드'], axis = 1, inplace = True)
#dust = dust.drop(['지역', '망', '측정소코드'], axis = 1)
```

```python
dust.head()
```

---

### 컬럼 이름 변경

```python
dust['city'] = '서울'
dust.head()
```

```python
dust.rename(columns = {'측정소명': 'name',
                      '측정일시': 'date',
                      '주소': 'addr'},
           inplace = True)
dust.columns
```

---

### 데이터 형(타입)변환

```python
dust.dtypes
```

```python
dust.head()
```

```python
dust['date'] = dust['date'].astype(str)
```

```python
dust['date'] = dust['date'].str.slice(0,8) #str.slice() : 문자열 일부분 짜르기
```

```python
dust.head()
```

```python
dust['date'] = pd.to_datetime(dust['date'])
dust.head()
```

```python
dust.dtypes
```

```python
dust['year'] = dust['date'].dt.year
dust['month'] = dust['date'].dt.month
dust['day'] = dust['date'].dt.day
dust.head()
```

### 데이터 병합하기

```python
s1 = pd.read_excel('nation.xlsx')
s2 = pd.read_excel('code.xlsx')
```

```python
s1
```

```python
s2
```

```python
pd.merge(s1, s2, on = '국적코드') #inner join
# on : 기준 변수
```

```python
pd.merge(s1, s2, on = '국적코드', how = 'outer') #outer join
#how = 'outer' : 공통 열이름 기준으로 합치되, 어느 한쪽에라도 없는 데이터가 있는 경우 NaN값이 지정된다.
```

```python
pd.merge(s1, s2, on = '국적코드', how = 'left')
#왼쪽에 입력한 데이터프레임 기준으로 각각의 key값에 해당하는 열을 지정된다.
```
