## 선(line) 그래프

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
xdata = [2, 4, 6, 8]
ydata = [1, 3, 5, 7]
```

### 그래프 제목 설정

```python
plt.plot(xdata, ydata)
plt.title('Line Graph')
plt.show()
```

### 다중 그래프

```python
data1 = [2, 4, 6, 8]
data2 = [8, 6, 4, 2]
plt.plot(data1, label = 'asc')
plt.plot(data2, label = 'desc')

plt.title('Line Graph')
plt.legend() #그래프내 범례 표현
plt.show()
```

### 그래프 색깔 변경

```python
plt.plot(xdata, ydata, color = 'red')
plt.title('Line Graph')
plt.show()
```

### 축 이름 설정

```python
plt.plot(xdata, ydata, color = 'red')
plt.title('Line Graph')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

### 그래프 스타일

```python
data1 = [2, 4, 6, 8]
data2 = [8, 6, 4, 2]
plt.plot(data1, color = 'r', label = 'dashed', linestyle = '--')
plt.plot(data2, color = 'g', label = 'dotted', linestyle = ':')

plt.title('Line Graph')
plt.legend() #그래프내 범주 표현
plt.show()
```

```python
xdata = [10, 20, 30, 40]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color = 'b', linestyle = '--', marker = 'o', markersize = '10')
```

### 그래프 범위 설정

```python
xdata = [10, 20, 30, 40]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color = 'b', linestyle = '--', marker = 'o', markersize = '10')
plt.xlim(0, 50)
plt.ylim(-5, 15)
plt.show()
```

```python
import pandas as pd
data = pd.read_excel('fine_dust.xlsx', index_col = 'area')
data
```

```python
data2020 = data[2020]
```

```python
plt.figure(figsize = (10, 4)) #figsize = (가로, 세로)
plt.plot(data2020, color = 'b', marker = 'o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2020 Fine Dust Line Graph')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (15, 6))

for year in range(2012, 2021):
    chartdata = data[year]
    plt.plot(chartdata, marker = 'o', label = year)

plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2012-2020 Fine Dust Line Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.style.use('bmh') # ggplot / Classic / Solarize_Light2 / default / bmh
for year in data.columns:
    plt.plot(data[year], marker = 'o', markersize = 5, label = year)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fine Dust 2012 - 2020', size=30)
plt.yticks([30,40,50], labels=['Good','soso','Bad'], size = 20)
plt.xticks(size = 20) # x축 눈금
plt.legend(loc = 'lower right', shadow = True ) # loc=[.88,.7]
plt.show()
```

## stripplot

```python
data6 = pd.read_excel('heatlh_screenings_2020.xlsx')
data6
```

```python
data6 = data6.loc[:, ['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]
data6
```

```python
data6.loc[data6['gender'] == 1, ['gender']] = 'male'
data6.loc[data6['gender'] == 2, ['gender']] = 'female'
```

```python
data6.loc[data6['drinking'] == 0, ['drinking']] = 'non-drinking'
data6.loc[data6['drinking'] == 1, ['drinking']] = 'drinking'
```

```python
data6.loc[data6['smoking'] == 1, ['smoking']] = 'non-smoking'
data6.loc[data6['smoking'] == 2, ['smoking']] = 'smoking'
```

```python
data6
```

```python
maledata = data6.loc[data6.gender == 'male', :]
femaledata = data6.loc[data6.gender == 'female', :]
```

```python
plt.figure(figsize = (10, 5))
plt.title('Seaborn Strip Plot Graph')

sns.stripplot(x='height', y='weight', data = maledata)
sns.stripplot(x='height', y='weight', data = femaledata)

plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.title('Seaborn Strip Plot Graph')

sns.stripplot(x='height', y='weight', hue = 'gender', data = maledata, palette='dark')
sns.stripplot(x='height', y='weight', hue = 'gender', data = femaledata, palette='Set1')

plt.show()
```

## violinplot 플럿

```python
plt.figure(figsize=(10, 6))
plt.title('Gender Weight Violin Plot Graph - Drinking category')

sns.violinplot(x='gender', y='weight', hue='drinking', data=data6[data6.weight < 120])
```
